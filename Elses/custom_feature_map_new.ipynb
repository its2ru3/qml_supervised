```python
import numpy as np
import matplotlib.pyplot as plt
import functools

from qiskit_aer import AerSimulator
from qiskit.circuit.library import ZZFeatureMap, PauliFeatureMap

from qiskit_machine_learning.kernels import FidelityQuantumKernel
from qiskit_machine_learning.algorithms import QSVC

from qiskit_machine_learning.datasets import ad_hoc_data
```

Backend

```python
backend = AerSimulator()
```

Dataset

```python
feature_dim = 2
sample_Total, training_dataset, test_dataset, class_labels = ad_hoc_data(
    training_size=10,
    test_size=5,
    n=feature_dim,
    gap=0.3
)
```

Helper: convert dataset format

```python
def dataset_to_array(dataset):
    X = []
    y = []
    for label, data in dataset.items():
        for x in data:
            X.append(x)
            y.append(label)
    return np.array(X), np.array(y)

X_train, y_train = dataset_to_array(training_dataset)
X_test, y_test = dataset_to_array(test_dataset)
```

🚀 Replace QSVM with QSVC + QuantumKernel

🔹 Example 1 (ZZFeatureMap)

```python
feature_map = ZZFeatureMap(feature_dimension=feature_dim, reps=2, entanglement='linear')

quantum_kernel = QuantumKernel(feature_map=feature_map, quantum_instance=backend)

model = QSVC(quantum_kernel=quantum_kernel)
model.fit(X_train, y_train)

accuracy = model.score(X_test, y_test)
print("testing success ratio:", accuracy)
```

Example 2 (default ZZFeatureMap)

```python
feature_map = ZZFeatureMap(feature_dimension=feature_dim, reps=2)

quantum_kernel = QuantumKernel(feature_map=feature_map, quantum_instance=backend)

model = QSVC(quantum_kernel=quantum_kernel)
model.fit(X_train, y_train)

print("testing success ratio:", model.score(X_test, y_test))
```

Custom data map (same as yours)

```python
def custom_data_map_func(x):
    coeff = x[0] if len(x) == 1 else functools.reduce(lambda m, n: m * n, np.pi - x)
    return coeff
```

ZZFeatureMap with custom mapping

```python
feature_map = ZZFeatureMap(
    feature_dimension=feature_dim,
    reps=2,
    data_map_func=custom_data_map_func
)

quantum_kernel = QuantumKernel(feature_map=feature_map, quantum_instance=backend)

model = QSVC(quantum_kernel=quantum_kernel)
model.fit(X_train, y_train)

print("testing success ratio:", model.score(X_test, y_test))
```

PauliFeatureMap

```python
feature_map = PauliFeatureMap(
    feature_dimension=feature_dim,
    reps=2,
    paulis=['Z','Y','ZZ']
)

quantum_kernel = QuantumKernel(feature_map=feature_map, quantum_instance=backend)

model = QSVC(quantum_kernel=quantum_kernel)
model.fit(X_train, y_train)

print("testing success ratio:", model.score(X_test, y_test))
```

Higher dimension case

```python
feature_dim = 3
sample_Total_b, training_dataset_b, test_dataset_b, class_labels = ad_hoc_data(
    training_size=20,
    test_size=10,
    n=feature_dim,
    gap=0.3,
    plot_data=False
)

X_train_b, y_train_b = dataset_to_array(training_dataset_b)
X_test_b, y_test_b = dataset_to_array(test_dataset_b)

feature_map = PauliFeatureMap(
    feature_dimension=feature_dim,
    reps=2,
    paulis=['Y','Z','ZZ','ZZZ'],
    data_map_func=custom_data_map_func
)

quantum_kernel = QuantumKernel(feature_map=feature_map, quantum_instance=backend)

model = QSVC(quantum_kernel=quantum_kernel)
model.fit(X_train_b, y_train_b)

print("testing success ratio:", model.score(X_test_b, y_test_b))
```
